## Funções aqui... Calcular o Lucro

```python
def calcular_lucro(preco_uni, custo_uni, quantidade_vendida):
    # Lucro Bruto = preco de venda e a quantidade vendida
    lucro_bruto = preco_uni * quantidade_vendida

    # Custo de venda = a quantidade vendia e o preco de custo do produto
    custo_vendido = custo_uni * quantidade_vendida

    #Lucro Liquido e a Renda
    lucro_liquido = lucro_bruto - custo_vendido

    #Lucro de venda por cada unidade
    lucro_por_unidade = preco_uni - custo_uni

    return lucro_liquido, lucro_por_unidade
```
