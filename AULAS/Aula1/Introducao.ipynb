## Variaveis

```python
Nome_produto = "Furadeira_Maquita 200"
Quantidade_estoque = 100
Custo_uni = 300
Preco_uni = 700
Quantidade_vendida = 20

#usando o snakecase
print("RELATORIO DE VENDAS DE FURADEIRAS")

Restante_estoque = Quantidade_estoque - Quantidade_vendida

print("Já foram vendidas: ", Quantidade_vendida,"\n","Restam no estoque: ", Restante_estoque)

```

```python
Custo_total = Quantidade_estoque * Custo_uni
# para saber o custo de tudo, pegar todos os itens no estoque e multiplicar pelo preco de cada um
print(" O custo total do estoque inicial foi de: ", Custo_total)
```

```python
# Como calcular a venda bruta, o quanto ele já vendeu
Retorno_bruto = Preco_uni * Quantidade_vendida
Custo_vendido = Quantidade_vendida * Custo_uni
Lucro_liquido = Retorno_bruto - Custo_vendido
print("Já tivemos um retorno bruto de: ", Retorno_bruto, "\n", "Tivemos um lucro de: ", Lucro_liquido )

#Outra forma de fazer
Lucro_liquido_por_produto = Preco_uni - Custo_uni
Lucro_liquido_otimizado = Quantidade_vendida * Lucro_liquido_por_produto
print("A cada unidade vendida lucramos (R$): ", Lucro_liquido_por_produto , "\n", "O lucro total foi de (R$): ", Lucro_liquido_otimizado)
```

```python

print("Coloque aqui o numero de vendas dessa semana: ")
Nova_venda = int(input("Vendas na semana:"))



if Quantidade_estoque <=0:
    print("Erro, valor no estoque abaixo de zero.")
elif Nova_venda > Restante_estoque:
    print("Erro: Ruptura de estoque. Não temos essa quantidade disponivel")
    print("Disponível atualemnte:")
    print(Restante_estoque)
    #elif Nova_venda >= 10:
    #    print("Você vendeu mais que 10")
else: 
    Restante_estoque = Restante_estoque - Nova_venda
    print("Restante estoque", Restante_estoque)
    Lucro_adicional = Nova_venda * (Preco_uni - Custo_uni)
    print("Lucro Adicional" , Lucro_adicional)
    Lucro_liquido = Lucro_liquido + Lucro_adicional
    print("Venda registrada! Novo estoque e novo lucro atualizado com sucesso")
    print(Lucro_liquido)

```

```python
while True:

    print("\n===================================")
    print("Comandos e funções:")
    print(" 'Mudar preco' = Alterar o preço de venda")
    print(" 'Custo' = Altera o custo unitário")
    print(" 'Add' = Adiciona novas vendas")
    print(" ' Sair' = Fecha o programa")
    print("\n====================================")

    comando = input("Digite o comando desejado: ").strip().lower()

    if comando == "4":
        print("Saindo do sistema de vendas.")
        break

    elif comando == "1" or comando == "mudar preco":
        Preco_uni = float(input(f"O preço atual é R${Preco_uni}. Digite o novo preço:"))
        print(f"Preço atualizado para R$ {Preco_uni}")
        if Preco_uni < Custo_uni:
            print("Voce nao pode vendar suas unidades abaixo do custo, vai ter prejuizo")
    
    elif comando == "2":
        Custo_uni = float(input(f"O custo atual é de R$ {Custo_uni}. Digite o novo custo: "))
        print(f"Custo atualizado para R$ {Custo_uni}")
    
    elif comando == "3":
        print("\n---REGISTRO DE NOVA VENDA ---")
        Nova_venda = int(input("Digite a quantidade de furadeiras vendidas agora (ou 0 para voltar): "))

        if Nova_venda == 0:
            print("Voltando ao menu principal.")
        elif Nova_venda < 0:
            print("ERRO: você não pode vender quantidades negativ. Operação anulada")
        elif Nova_venda > Restante_estoque:
            print("ERRO: Overbooking de estoque. Não temos essa quantidade disponível.")
            print("Disponivel atualmente" , Restante_estoque)
        else:
            Restante_estoque = Restante_estoque - Nova_venda
            print("Restante estoque", Restante_estoque)
            Lucro_adicional = Nova_venda * (Preco_uni - Custo_uni)
            print("Lucro Adicional" , Lucro_adicional)
            Lucro_liquido = Lucro_liquido + Lucro_adicional
            print("Venda registrada! Novo estoque e novo lucro atualizado com sucesso")
            print(Lucro_liquido)
    
    else: 
        print("Comando invalido tente novamente.")


    
```
