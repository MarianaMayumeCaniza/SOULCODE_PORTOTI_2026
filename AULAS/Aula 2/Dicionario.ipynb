```python
#Lista da Lanchonete

import datetime

estoque = [
    {"nome": "Cachorro-Quente" , "preco": 12.50,"quantidade": 20 },
    {"nome": "Refri Lata" , "preco": 6.00,"quantidade": 30 }

]

historico_vendas =[]
caixa = 0.0 #controle de caixa

while True:


    print(f"""    
    ===================================
    LANCHONETE SOUL

    Caixa atual: R# {caixa}
    --------------------------


    """)
    print("CARDAPIO")

    for id_produto, produto in enumerate(estoque):
        print(f"[{id_produto}] {produto['nome']} | R$ {produto['preco']} | Estoque: {produto['quantidade']} ")
    

    print("""
    ===================================
    1 = Registra venda
    2 = Cadastra novo produto
    3 = Fechar caixa
    4 = Alterar Preço de um Produto
    5 = Adicionar Produtos ao Estoque
    6 = Buscar produto
    7 = Promoções
    8 = Nota fiscal
    
    """)
    
    comando = input("Digite o comando: ").strip()

    if comando == "3":
        print(f"Encerrando caixa. Total do caixa R${caixa}")
        break
    
    elif comando == "1":
        id_venda = int(input("Digite o numero [ID] do produto vendido: "))
        print((estoque[0]['quantidade']))
        if id_venda < 0 or id_venda>=len(estoque):
            print("ERRO: ID invalido, SE ATENTE AOS COMANDOS!!!!")
        
        else: 
            qtd = int(input(f"Quantos {estoque[id_venda]['nome']} foram vendidos"))

            if qtd <= 0:
                print("ERRO: QUANTIDADE INVALIDA")
            elif qtd > estoque[id_venda]['quantidade']:
                print(f" ESTOQUE INSUFICIENTE: Restam {estoque[id_venda]['quantidade']}")
            else:
                estoque[id_venda]['quantidade'] = estoque[id_venda]['quantidade'] - qtd
                valor_total = qtd * estoque[id_venda]['preco']
                caixa += valor_total

                #REGISTRO DE DATA E HORA RECIBO
                historico_vendas.append({
                    "horarios": datetime.datetime.now().strftime("%H:%M:%S"),
                    "item": estoque[id_venda]['nome'],
                    "qtd": qtd,
                    "valor": valor_total
                })
                print(f"Venda registrada com sucesso! Entrou no caixa R$ {valor_total}")
    
    elif comando == "2":
        print("""
            ------------- Cadastrar novo produto -----------------


        """)

        novo_nome = input("Qual o nome dos produtos? ")
        novo_preco = float(input("Qual o preço de venda? R$"))
        nova_qtd = int(input("Qual a quantidade incial do produto? "))

        # Adicionando o novo produto a lista
        novo_produto = {
            "nome": novo_nome,
            "preco": novo_preco,
            "quantidade": nova_qtd
        }

        #Adicionando ao estoque
        estoque.append(novo_produto)
        print(f"Produto {novo_nome} adicionado ao sistema")
    
    elif comando =="4":
        print("""
        Alterar preço do produto""")

        id_produto = int(input(f" Digite o [ID] do produto: "))

        if id_produto < 0 or id_produto >= len(estoque):
            print("ERRO: ID Invalido")
        else:
            novo_preco = float (input(f""" Digite o novo preço do {estoque[id_produto]['nome']}
            o atual preço é R$ {estoque[id_produto]['preco']} """))

            estoque[id_produto]['preco'] = novo_preco
            print(f"""O preço do produto {estoque[id_produto]['nome']} foi alterado para:{estoque[id_produto]['preco']}""")
            
    
    elif comando == "5":
        print("---------------- ADICIONAR AO ESTOQUE ------------------------")
        id_produto = int(input("Digite o [ID] do produto que deseja repor: "))

        if id_produto <0 or id_produto >= len(estoque):
            print("ERRO: ID invalido")
        else: 
            qtd_reposicao = int(input(f"Quantas unidade deseja de {estoque[id_produto]['nome']}"))

            if qtd_reposicao <= 0:
                print("ERRO: QUANTIDADE INVALIDA")
            else: 
                estoque[id_produto]['quantidade'] += qtd_reposicao
                print(f" SUCESSO! Agora existem {estoque[id_produto]['quantidade']} unidades de {estoque[id_produto]['nome']} no estoque.")
    

    elif comando == "6":
        print("\n -------- Pesquisar Produto ------------")
        termo_busca = input("Digite o nome do produto:").strip().lower()

        encontrou = False

        for produto in estoque:
            if termo_busca in produto['nome'].lower():
                print(f"RESULTADOS DA PESQUISA: {produto['nome']} | R$ {produto['preco']} | Estoque: {produto['quantidade']}")
                encontrou = True
        if not encontrou:
            print("Nenhum resultado encontrado")
    
    elif comando == "7":
        print("----------- PROMOÇÕES ---------------------")
        print("1 - Aplicar desconto em 1(UM) unico produto especifico")
        print("2 - Aplicar desconto em TODO o cardapio")
        
        tipo_promo = input("Escolha  o tipo do desconto: ").strip()

        porcentagem = float(input("Qual a porcentagem(%) do desconto que será adicionado: "))

        fator_desconto = (100 - porcentagem)/100

        if tipo_promo == "1":
            id_produto = int(input("Digite o [ID] do produto que recebera o desconto: "))
            if id_produto < 0 or id_produto >= len(estoque):
                print("ERRO [ID] INVALIDO")
            else: 
                preco_antigo = estoque[id_produto]['preco'] 
                estoque[id_produto]['preco']= round(preco_antigo * fator_desconto, 2)
                print(f"SUCESSO!! {estoque[id_produto]['nome']} mudou de R$ {preco_antigo} para {estoque[id_produto]['preco']}")

        elif tipo_promo == "2":
            for produto in estoque:
                produto['preco'] = round(produto['preco'] * fator_desconto, 2)
            print(f"SUCESSO!!! Você deu um desconto de {porcentagem} A TODOS OS PRODUTOS do cardapio")
        else: 
            print("COMANDO INVALIDO")
    
    elif comando == "8":
        print("----------------- RECIBO DE VENDAS ----------------")
        if len(historico_vendas) == 0:
            print("Nenhuma venda ainda.")
        else: 
            for venda in historico_vendas: 
                print(f"[{venda['horarios']}] {venda['item']} x{venda['qtd']}")







    
    

    





```
