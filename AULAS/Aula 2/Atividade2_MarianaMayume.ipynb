LIÇÃO DE CASA

recriar empresa EXEMPLO que tenha pelo menos  10 produtos
tenha menu de interface, menu intuitivo, bem descritivo,
tem q ser possivel: adicionar ao estoque, alterar preço de produto
registrar venda, emitir recibo(tem q ter pelo menos, data/hora e quantos foram vendidos do produto), pesquisar por nome, fazer promoções, fechar opção, usar pelo menos uma vez cada no codigo os comandos if, else, elif, if not, try & except, o uso apropriado 

```python
import datetime

estoque = [
    {"nome": "Cafe",            "preco": 4.50,  "quantidade": 80},
    {"nome": "Chocolate Quente","preco": 8.00,  "quantidade": 60},
    {"nome": "Cappuccino",      "preco": 9.50,  "quantidade": 50},
    {"nome": "Cha Matte",       "preco": 5.00,  "quantidade": 40},
    {"nome": "Latte",           "preco": 10.00, "quantidade": 35},
    {"nome": "Descafeinado",    "preco": 5.50,  "quantidade": 45},
    {"nome": "Frapuccino",      "preco": 12.00, "quantidade": 55},
    {"nome": "Leite",           "preco": 3.50,  "quantidade": 70},
    {"nome": "Mocca",           "preco": 11.00, "quantidade": 25},
    {"nome": "Cafe com Leite",  "preco": 6.00,  "quantidade": 65},
]

historico_vendas =[]
caixa = 0.0 #controle de caixa


# ============================================
# FUNÇÕES
# ============================================

def exibir_cardapio():
    print("\n CARDAPIO CAFETERIA:")
    for id_produto, produto in enumerate(estoque):
        print(f"[{id_produto}] {produto['nome']} | R$ {produto['preco']} | Estoque: {produto['quantidade']} ")



while True:

    print(f"""
    ====================================
    CAFETERIA

    Caixa atual: R$ {caixa:.2f}
    ====================================""")
        
    exibir_cardapio()

    # ============================================
    # OPCOES
    # ============================================

    print("""
    ====================================
    Escolha uma opção abaixo:

    1 - Registrar venda
    2 - Cadastrar novo produto
    3 - Fechar caixa
    4 - Alterar preço de produto
    5 - Repor estoque
    6 - Buscar produto
    7 - Promoções
    8 - Emitir recibo
    ====================================

    Digite 0 para sair do sistema

    """)
    comando = input("Digite um número do sistema para executar a ação: ").strip()

    # 1 REGISTRAR VENDA =================================
    if comando == "1":
        print("\n-------- REGISTRAR VENDA --------")
        try:
            id_venda = int(input("Digite o [ID] do produto vendido: "))

            if id_venda < 0 or id_venda >= len(estoque):
                print("ERRO: ID inválido! [ID] Inexistente, escolha um item do cardápio para registrar a venda")
                exibir_cardapio()
            else:
                produto = estoque[id_venda]
                qtd = int(input(f"Quantos '{produto['nome']}' foram vendidos? "))
                if qtd <= 0:
                    print("ERRO: quantidade inválida, negativa. Digite uma quantidade válida")
                elif qtd > produto['quantidade']:
                    print(f"ERRO: estoque insuficiente! Restam {produto['quantidade']} unidades.")
                else:
                    produto['quantidade'] -= qtd
                    valor_total = qtd * produto['preco']
                    caixa += valor_total

                    historico_vendas.append({
                        "data":    datetime.datetime.now().strftime("%d/%m/%Y"),
                        "horario": datetime.datetime.now().strftime("%H:%M:%S"),
                        "item":    produto['nome'],
                        "qtd":     qtd,
                        "valor":   valor_total
                    })
                    print(f"Venda registrada! Entrou no caixa: R$ {valor_total:.2f}")
        except ValueError:
            print("ERRO: digite apenas números do menu!")

    # 2 CADASTRAR PRODUTO =================================
    elif comando == "2":
        print("\n-------- CADASTRAR PRODUTO --------")
        
        novo_nome  = input("Digite o nome do produto: ")
        novo_preco = float(input("Qual será o preço de venda: R$ "))
        nova_qtd   = int(input("Quantidade inicial do estoque: "))

        if novo_preco <= 0 or nova_qtd <= 0:
            print("ERRO: preço e quantidade devem ser maiores que zero. Digite uma quantidade valida")
        else:
            novo_cafe = {
            "nome": novo_nome,
            "preco": novo_preco,
            "quantidade": nova_qtd
            }

            estoque.append(novo_cafe)
            print(f"Produto '{novo_nome}' cadastrado com sucesso!")


    # 3 FECHAR CAIXA =================================
        if comando == "3":
            print(f"\nEncerrando caixa. Total do dia: R$ {caixa:.2f}")
            print("""
            Saindo............................

            Até amanhã! """)

            break

    # 4 ALTERAR PREÇO =================================
    elif comando == "4":
        print("\n-------- ALTERAR PREÇO --------")
        try:
            id_produto = int(input("Digite o [ID] do produto: "))

            if id_produto < 0 or id_produto >= len(estoque):
                print("ERRO: ID inválido!")
            else:
                produto = estoque[id_produto]
                novo_preco = float(input(f"Preço atual de '{produto['nome']}': R$ {produto['preco']:.2f}\n Qual será o novo preço (R$) ? "))

                if novo_preco <= 0:
                    print("ERRO: preço inválido. Voce precisa digitar um valor maior que 0")
                else:
                    produto['preco'] = novo_preco
                    print(f"Preço de '{produto['nome']}' atualizado para R$ {novo_preco:.2f}")

        except ValueError:
            print("Digite um [ID] de produto valido!")
            exibir_cardapio()

    # 5 REPOR ESTOQUE =================================
    elif comando == "5":
        print("\n-------- REPOR ESTOQUE --------")
        try:
            id_produto = int(input("Digite o [ID] do produto: "))

            if id_produto < 0 or id_produto >= len(estoque):
                print("ERRO: ID inválido!")
            else:
                produto = estoque[id_produto]
                qtd_reposicao = int(input(f"Quantas unidades deseja adicionar ao '{produto['nome']}'? "))

                if qtd_reposicao <= 0:
                    print("ERRO: quantidade inválida. Digite um numero maior que 0")
                else:
                    produto['quantidade'] += qtd_reposicao
                    print(f"Estoque reposto! '{produto['nome']}' agora tem {produto['quantidade']} unidades.")

        except ValueError:
            print("Digite um [ID] de produto valido!")
            exibir_cardapio()

    # 6 BUSCAR PRODUTO =================================
    elif comando == "6":
        print("\n-------- PESQUISAR PRODUTO --------")
        termo_busca = input("Digite o nome do produto: ").strip().lower()

        resultados  = [p for p in estoque if termo_busca in p['nome'].lower()]

        encontrou = False

        for produto in estoque:
            if termo_busca in produto['nome'].lower():
                print(f"RESULTADOS DA PESQUISA: {produto['nome']} | R$ {produto['preco']} | Estoque: {produto['quantidade']}")
                encontrou = True
        if not encontrou:
            print("Nenhum resultado encontrado")

    # 7 PROMOÇÕES =================================
    elif comando == "7":

        print("\n-------- PROMOÇÕES --------")
        print("  1 - Desconto em um produto específico")
        print("  2 - Desconto em todo o cardápio")

        tipo_promo = input("Escolha o tipo: ").strip()

        porcentagem = float(input("Qual a porcentagem(%) do desconto que será adicionado: "))

        fator_desconto = (100 - porcentagem)/100

        if tipo_promo == "1":
            id_produto = int(input("Digite o [ID] do produto que recebera o desconto: "))
            if id_produto < 0 or id_produto >= len(estoque):
                print("ERRO [ID] INVALIDO")
            else: 
                preco_antigo = estoque[id_produto]['preco'] 
                estoque[id_produto]['preco']= round(preco_antigo * fator_desconto, 2)
                print(f"SUCESSO!! {estoque[id_produto]['nome']} mudou de R$ {preco_antigo} para {estoque[id_produto]['preco']}")

        elif tipo_promo == "2":
            for produto in estoque:
                produto['preco'] = round(produto['preco'] * fator_desconto, 2)
            print(f"SUCESSO!!! Você deu um desconto de {porcentagem} A TODOS OS PRODUTOS do cardapio")
        else:
            print("COMANDO INVALIDO")

    # 8 RECIBO =================================
    elif comando == "8":
        print("\n======= RECIBO DE VENDAS =======")

        if not historico_vendas:
            print("Nenhuma venda registrada ainda.")
        else:
            for venda in historico_vendas:
                print(f"[{venda['horarios']}] {venda['item']} x{venda['qtd']}")
        print("================================")
    
    elif comando == "0":
        print("Saindo ......................................")
        break

    # ── COMANDO INVÁLIDO ────────────────────
    else:
        print("Comando inválido! Tente novamente.")


```
